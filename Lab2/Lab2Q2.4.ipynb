```python
%matplotlib notebook

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib import cm
from math import e
import io


def calcSy(mu, nu):
    Sx = 1/12
    a = 1 - 0.99*(np.cos(mu) + np.cos(nu)) + 0.9801*(np.cos(mu + nu))
    b = 0.99*(np.sin(mu) + np.sin(nu)) - 0.9801*(np.sin(mu + nu))
    k = 3/(a**2 + b**2)
    absH = k*np.sqrt(a**2 + b**2)
    Sy = np.log((absH**2)*Sx)
    return Sy
```

```python
N = 64
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
mu = np.linspace(-np.pi,np.pi,num=N)
nu = np.linspace(-np.pi,np.pi,num=N)

X, Y = np.meshgrid(mu, nu)
Z = calcSy(X, Y)
```

```python
surf = ax.plot_surface(X, Y, Z, cmap=plt.cm.coolwarm)

ax.set_xlabel('$\mu$ axis')
ax.set_ylabel('$\\nu$ axis')
ax.set_zlabel('Z Label')

fig.colorbar(surf, shrink=0.5, aspect=5, pad = 0.1)
```

```python
import io
# Save the image in memory in PNG format
png1 = io.BytesIO()
fig.savefig(png1, format="png")

# Load this image into PIL
img_out = Image.open(png1)

# Save as TIFF
img_out.save("L2Q2Theoretical.tif")
png1.close()

plt.show()

print(Z[32][32])
```

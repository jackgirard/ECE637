```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from PIL import Image
from math import e
import io

def H(mu, nu):
    return abs((1/25)*(1 + 2*(np.cos(2*mu)+np.cos(mu)))*(1 + 2*(np.cos(2*nu)+np.cos(nu))))

def G(mu, nu, lam):
    return abs(1 + lam*(1 - H(mu, nu)))
```

```python
mu = np.linspace(-np.pi,np.pi,180)
nu = np.linspace(-np.pi,np.pi,180)

X, Y = np.meshgrid(mu, nu)
```

```python
Z = H(X, Y)
fig1 = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$\nu$')
ax.set_zlabel(r'$H$')
ax.set_title('Magnitude of Frequency Response ' + r'$|H(e^{j\mu}, e^{j\nu})|$' + ' vs ' + r'$\mu$' + ' and ' + r'$\nu$')
fig1.set_figwidth(8)
fig1.set_figheight(8)
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor=(0,0,0))
plt.show()
```

```python
lam = 1.5
Z = G(X, Y, lam)
fig2 = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$\nu$')
ax.set_zlabel(r'$H$')
ax.set_title(r'$|G(e^{j\mu}, e^{j\nu})|$' + ' vs ' + r'$\mu$' + ' and ' + r'$\nu$' + ' at ' + r'$\lambda = 1.5$')
fig2.set_figwidth(8)
fig2.set_figheight(8)
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor=(0,0,0))
plt.show()
```

```python
# Save the image in memory in PNG format
png1 = io.BytesIO()
png2 = io.BytesIO()
fig1.savefig(png1, format="png")
fig2.savefig(png2, format="png")

# Load this image into PIL
img_out1 = Image.open(png1)
img_out2 = Image.open(png2)

# Save as TIFF
img_out1.save("HPlotQ4.tif")
img_out2.save("GPlotQ4.tif")
png1.close()
png2.close()
```

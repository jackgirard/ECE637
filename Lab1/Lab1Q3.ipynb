```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
from PIL import Image
from math import e
import io

def H(mu, nu):
    return abs((1/81)*(1 + 2*(np.cos(4*mu)+np.cos(3*mu)+np.cos(2*mu)+np.cos(mu)))*(1 + 2*(np.cos(4*nu)+np.cos(3*nu)+np.cos(2*nu)+np.cos(nu))))
```

```python
mu = np.linspace(-np.pi,np.pi,180)
nu = np.linspace(-np.pi,np.pi,180)

X, Y = np.meshgrid(mu, nu)
Z = H(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel(r'$\mu$')
ax.set_ylabel(r'$\nu$')
ax.set_zlabel(r'$H$')
ax.set_title('Magnitude of Frequency Response ' + r'$|H(e^{j\mu}, e^{j\nu})|$' + ' vs ' + r'$\mu$' + ' and ' + r'$\nu$')
fig.set_figwidth(8)
fig.set_figheight(8)
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor=(0,0,0))
plt.show()
```

```python
# Save the image in memory in PNG format
png1 = io.BytesIO()
fig.savefig(png1, format="png")

# Load this image into PIL
img_out = Image.open(png1)

# Save as TIFF
img_out.save("3dPlotQ3.tif")
png1.close()
```
